A signal is divided into a number of different windows and Fourier transforms for these windows are computed. When frequency components vary over time in a signal, STFT provides the time localized information of these variations. On the other hand a Fourier transform just provides frequency information avergaed over the entire time duration. 

The STFT elements and their inverse are computed as follows - 




$
\chi_{STFT} [m, n]= \sum\limits^{L-1}_{N=0} {x[k] g[k-m] e^{-\frac{j2 \pi kn}{L}}}
$

$
x[k] = \sum\limits_{m}\sum\limits_{n} {\chi_{STFT} [m, n] g[k-m] e^{\frac{j2 \pi kn}{L}}}
$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
from scipy import signal

def generate_signal(type, duration=5.0, fs=1000, freq=50):
    t = np.linspace(0, duration, int(fs*duration), endpoint=False)
    if type == 'sine':
        return np.sin(2 * np.pi * freq * t), t
    elif type == 'square':
        return signal.square(2 * np.pi * freq * t), t
    elif type == 'sawtooth':
        return signal.sawtooth(2 * np.pi * freq * t), t

def perform_stft(signal, fs):
    f, t, Zxx = stft(signal, fs=fs, nperseg=512)
    return f, t, np.abs(Zxx)

def plot_signal_and_spectrogram(signal, t, fs):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(211)
    plt.plot(t, signal)
    plt.title('Time Domain Signal')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')

    plt.subplot(212)
    f, t, Zxx = perform_stft(signal, fs)
    plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
    plt.title('STFT Magnitude')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [s]')
    plt.tight_layout()
    plt.show()

sine_signal, t = generate_signal('sine')
plot_signal_and_spectrogram(sine_signal, t, 1000)


```

```python
square_signal, t = generate_signal('square')
plot_signal_and_spectrogram(square_signal, t, 1000)
```

```python

sawtooth_signal, t = generate_signal('sawtooth')
plot_signal_and_spectrogram(sawtooth_signal, t, 1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

def perform_stft_librosa(audio_path, n_fft=2048, hop_length=512):
    y, sr = librosa.load(audio_path, sr=None)
    D = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

    plt.figure(figsize=(12, 6))
    librosa.display.specshow(S_db, sr=sr, hop_length=hop_length, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('STFT Spectrogram')
    plt.show()

def compare_stft_windows(audio_path, n_fft_high, n_fft_low, hop_length=512):
    y, sr = librosa.load(audio_path, sr=None)
    
    # Compute STFT for higher window size
    D_high = librosa.stft(y, n_fft=n_fft_high, hop_length=hop_length)
    S_db_high = librosa.amplitude_to_db(np.abs(D_high), ref=np.max)
    
    # Compute STFT for lower window size
    D_low = librosa.stft(y, n_fft=n_fft_low, hop_length=hop_length)
    S_db_low = librosa.amplitude_to_db(np.abs(D_low), ref=np.max)
    
    # Plot spectrograms
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    librosa.display.specshow(S_db_high, sr=sr, hop_length=hop_length, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('STFT Spectrogram (Higher Window Size)')
    
    plt.subplot(1, 2, 2)
    librosa.display.specshow(S_db_low, sr=sr, hop_length=hop_length, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('STFT Spectrogram (Lower Window Size)')
    
    plt.tight_layout()
    plt.show()

audio_path = 'Free-birds-chirping-sound-effect.wav'
compare_stft_windows(audio_path, n_fft_high=8192, n_fft_low=256)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
from scipy.io import wavfile
import librosa
import librosa.display

def perform_stft_librosa(audio_path):
    # Load the audio file using scipy
    sr, y = wavfile.read(audio_path)

    # Normalize audio data
    y = y.astype(np.float32) / np.max(np.abs(y))

    # Apply low-pass filters using different methods
    cutoff_freq = 5000  # Adjust cutoff frequency as needed
    nyquist_freq = sr / 2
    order = 5  # Filter order

    # Method 1: Butterworth filter
    b, a = signal.butter(order, cutoff_freq / nyquist_freq, btype='low')
    y_filtered_butter = signal.filtfilt(b, a, y)

    # Method 2: FIR filter design
    taps_fir = signal.firwin(order * 2 + 1, cutoff_freq / nyquist_freq, window='hamming')
    y_filtered_fir = signal.convolve(y, taps_fir, mode='same')

    # Method 3: Direct truncation in frequency domain (ideal low-pass filter)
    Y = np.fft.rfft(y)
    freqs = np.fft.rfftfreq(len(y), d=1/sr)
    Y[np.abs(freqs) > cutoff_freq] = 0
    y_filtered_freq_domain = np.fft.irfft(Y)

    # Perform STFT on original and filtered signals
    D_original = librosa.stft(y, n_fft=2048, hop_length=512)
    D_butter = librosa.stft(y_filtered_butter, n_fft=2048, hop_length=512)
    D_fir = librosa.stft(y_filtered_fir, n_fft=2048, hop_length=512)
    D_freq_domain = librosa.stft(y_filtered_freq_domain, n_fft=2048, hop_length=512)

    # Plot spectrograms for comparison
    plt.figure(figsize=(12, 10))
    plt.subplot(4, 1, 1)
    librosa.display.specshow(librosa.amplitude_to_db(np.abs(D_original), ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Original Signal')

    plt.subplot(4, 1, 2)
    librosa.display.specshow(librosa.amplitude_to_db(np.abs(D_butter), ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Butterworth Filter')

    plt.subplot(4, 1, 3)
    librosa.display.specshow(librosa.amplitude_to_db(np.abs(D_fir), ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('FIR Filter')

    plt.subplot(4, 1, 4)
    librosa.display.specshow(librosa.amplitude_to_db(np.abs(D_freq_domain), ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Frequency Domain Truncation')

    plt.tight_layout()
    plt.show()

    # Write filtered audio files
    wavfile.write('filtered_audio_butter.wav', sr, (y_filtered_butter * 32767).astype(np.int16))
    wavfile.write('filtered_audio_fir.wav', sr, (y_filtered_fir * 32767).astype(np.int16))
    wavfile.write('filtered_audio_freq_domain.wav', sr, (y_filtered_freq_domain * 32767).astype(np.int16))

# Example usage
audio_path = 'StarWars60.wav'
perform_stft_librosa(audio_path)
```

```python
from transformers import ASTFeatureExtractor

feature_extractor = ASTFeatureExtractor()
```

```python
import torchaudio

waveform, sampling_rate = torchaudio.load('sample1.wav')
waveform = waveform.squeeze().numpy()

waveform = waveform.reshape(-1)
waveform.shape
```

```python
inputs = feature_extractor(waveform, sampling_rate=sampling_rate, padding="max_length", return_tensors="pt")
input_values = inputs.input_values
print(input_values.shape)     
```

```python
from transformers import AutoModelForAudioClassification

model = AutoModelForAudioClassification.from_pretrained("MIT/ast-finetuned-audioset-10-10-0.4593")
```

```python
import torch

with torch.no_grad():
  outputs = model(input_values)
```

```python
predicted_class_idx = outputs.logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
